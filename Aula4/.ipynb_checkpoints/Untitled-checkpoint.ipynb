```python
import cv2
from matplotlib import pyplot as plt
```

```python
image = cv2.imread("playing-cards.jpg", cv2.IMREAD_GRAYSCALE)
plt.imshow(image)
```

```python
image_template = cv2.imread("playing-cards-7-copas.jpg", cv2.IMREAD_GRAYSCALE)
plt.imshow(image_template)
```

```python
#image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image_matched = image.copy()
matched_template = cv2.matchTemplate(image, image_template, cv2.TM_CCOEFF_NORMED)
w, h = image_template.shape[::-1]
```

```python
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(matched_template)
top_left = max_loc
bottom_right = (top_left[0] + w, top_left[1] + h)
cv2.rectangle(image_matched, top_left, bottom_right, (0,255,0), 2)
```

```python
plt.imshow(image_matched)
```
